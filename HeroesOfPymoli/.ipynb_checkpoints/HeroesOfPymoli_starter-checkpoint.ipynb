```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

# Cleaning data from non valid values
clean_purchase_data = purchase_data.dropna(how="any")
clean_purchase_data.head()
```

## Player Count

* Display the total number of players


```python

Total_number_of_players = len(pd.unique(clean_purchase_data['SN']))
Player_Count_df = pd.DataFrame({"Total number of players" : [Total_number_of_players]})

Player_Count_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
Number_of_Unique_Items = len(pd.unique(clean_purchase_data['Item ID']))
Purchasing_Analysis_df = pd.DataFrame({
    "Number of Unique Items" : [Number_of_Unique_Items],
    "Average Price" : [clean_purchase_data['Price'].mean()],
    "Number of Purchases" : [len(pd.unique(clean_purchase_data['Purchase ID']))],
    "Total Revenue" : [clean_purchase_data['Price'].sum()]
                    })

Purchasing_Analysis_Formatted_df = Purchasing_Analysis_df.style.format({'Number of Unique Items': '{:,.0f}', 'Average Price':'${:,.2f}', 'Number of Purchases':'{:,.0f}', 'Total Revenue':'${:,.2f}'})
# Show Purchasing_Analysis_Formatted_df
Purchasing_Analysis_Formatted_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Remove duplicates in source dataframe
clean_purchase_data_no_duplicated_players = clean_purchase_data.drop_duplicates(subset=['SN'])
# clean_purchase_data_no_duplicated_players.head()
```

```python
# Group by 'SN', 'Gender' and calculate Total and Percent
Gender_Demographics_df = pd.DataFrame( clean_purchase_data_no_duplicated_players.groupby(['SN','Gender'])['Gender'].count().groupby(['Gender']).count())

# Rename to Total Count
Gender_Demographics_df = Gender_Demographics_df.rename(columns={'Gender':"Total Count"})

# Order by Total Count
Gender_Demographics_df = Gender_Demographics_df.sort_values(by=['Total Count'], ascending=False)

# Create "Percentage of Players" column in Gender_Demographics_df
Gender_Demographics_df["Percentage of Players"] = Gender_Demographics_df['Total Count'] / Total_number_of_players * 100

# Format "Percentage of Players" column as float
Gender_Demographics_df["Percentage of Players"] = Gender_Demographics_df["Percentage of Players"].astype(float).map("{:,.2f}%".format)

# Show Gender_Demographics_df
Gender_Demographics_df
```


## Purchasing Analysis (Gender)

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Create the bins in which Data will be held
# Bins are 0, 9.9, 14, 19, 24, 29, 34, 39, 200.   
bins = [0, 9.9, 14, 19, 24, 29, 34, 39, 200]

# Create the names for the five bins
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
clean_purchase_data["Age range"] = pd.cut(clean_purchase_data["Age"], bins, labels=group_names, include_lowest=True)
# clean_purchase_data.head(5)
```

```python
#Remove duplicates 'SN' in clean_purchase_data
Age_Demographics_unique_sn_df = clean_purchase_data.drop_duplicates(subset='SN')

#Group by Age range and count players
Age_Demographics_df = (Age_Demographics_unique_sn_df.groupby(['Age range']).agg({'SN': [np.ma.count]}).rename(columns={'count': 'Total Count'})) 

# Rename columns
Age_Demographics_df = Age_Demographics_df.rename(columns={'SN':"Total Count"})

#Add Percentage of Players column
Age_Demographics_df["Percentage of Players"] = Age_Demographics_df['Total Count'] / Total_number_of_players * 100

#Format Percentage of Players column
Age_Demographics_df["Percentage of Players"] = Age_Demographics_df["Percentage of Players"].astype(float).map("{:,.2f}%".format)

# Show dataframe
Age_Demographics_df.head(100)

```

```python

# Calculate 
Purchasing_Analysis_Age["Average Purchase Price"] = Purchasing_Analysis_Age['Price', 'Total Purchase Value'].astype(float) / Purchasing_Analysis_Age['Purchase ID',       'Purchase Count'].astype(float)
Purchasing_Analysis_Age["Avg Total Purchase per Person"] =Purchasing_Analysis_Age['Price', 'Total Purchase Value'].astype(float) / Purchasing_Analysis_Age[         'SN',              'nunique'].astype(float)

#Format Percentage of Players column
Purchasing_Analysis_Age["Average Purchase Price"] = Purchasing_Analysis_Age["Average Purchase Price"].astype(float).map("${:,.2f}".format)
Purchasing_Analysis_Age["Avg Total Purchase per Person"] = Purchasing_Analysis_Age["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)

Purchasing_Analysis_Age.head(100)
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Group by SN and count players and sum Price
                # Top_Spenders_df = (clean_purchase_data.groupby(['SN']).agg({'Purchase ID': [np.ma.count], 'Price': [np.sum]}).rename(columns={'count': 'Purchase Count', 'sum': 'Total Purchase Value'})) 
Top_Spenders_df = (clean_purchase_data.groupby(['SN']).agg(Purchase_Count=pd.NamedAgg(column='SN', aggfunc='count'), Total_Purchase_Value=pd.NamedAgg(column='Price', aggfunc=sum))) 

# Add Average Purchase Price column  
Top_Spenders_df["Average Purchase Price"] = Top_Spenders_df["Total_Purchase_Value"] / Top_Spenders_df["Purchase_Count"]

#Format Percentage of Players column
Top_Spenders_df["Average Purchase Price"] = Top_Spenders_df["Average Purchase Price"].astype(float).map("${:,.2f}".format)
Top_Spenders_df["Total_Purchase_Value"] = Top_Spenders_df["Total_Purchase_Value"].astype(float).map("${:,.2f}".format)

# Sort descending Purchase_count
Top_Spenders_df = Top_Spenders_df.sort_values(by=['Purchase_Count'], ascending=False)

Top_Spenders_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
clean_purchase_data.head()
```

```python
#Group by "Item ID" and count players and sum Price
Most_Popular_Items_df = (clean_purchase_data.groupby(['Item ID','Item Name']).agg(Purchase_Count=pd.NamedAgg(column='Item ID', aggfunc='count'), Total_Purchase_Value=pd.NamedAgg(column='Price', aggfunc=sum))) 

# Add Average Purchase Price column  
Most_Popular_Items_df["Item Price"] =  Most_Popular_Items_df["Total_Purchase_Value"] / Most_Popular_Items_df["Purchase_Count"] 

#Format Percentage of Players column
Most_Popular_Items_df["Item Price"] = Most_Popular_Items_df["Item Price"].astype(float).map("${:,.2f}".format)
Most_Popular_Items_df["Total_Purchase_Value"] = Most_Popular_Items_df["Total_Purchase_Value"].astype(float).map("${:,.2f}".format)

# Sort descending Purchase_count
Most_Popular_Items_df = Most_Popular_Items_df.sort_values(by=['Purchase_Count'], ascending=False)

Most_Popular_Items_df
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
# Create a copy of Most_Popular_Items_df to avoid modify the source dataframe
Most_Profitable_Items_raw_df = Most_Popular_Items_df.copy()

# Clean column Total_Purchase_Value removing $ sign in order to sort by this column as a number
Most_Profitable_Items_raw_df['Total_Purchase_Value'] = Most_Profitable_Items_raw_df['Total_Purchase_Value'].str.replace('$','').astype(float)

# Sort the above table by total purchase value in descending order
Most_Profitable_Items_df = Most_Profitable_Items_raw_df.sort_values(by=['Total_Purchase_Value'], ascending=False)

# Optional: give the displayed data cleaner formatting
Most_Profitable_Items_df["Total_Purchase_Value"] = Most_Profitable_Items_df["Total_Purchase_Value"].map("${:,.2f}".format)

# Display a preview of the data frame
Most_Profitable_Items_df.head()


```
